## 1. Contexte général

La conduite automobile est une activité complexe qui mobilise plusieurs ressources cognitives : perception visuelle, attention soutenue, prise de décision, anticipation, contrôle moteur et gestion des imprévus.

Le papier CL-Drive s'intéresse donc à une question centrale :

> Peut-on estimer automatiquement la charge cognitive d'un conducteur à partir de signaux physiologiques et comportementaux mesurés pendant la conduite ?

#### Q1. Pourquoi est-il intéressant d'estimer la charge cognitive d'un conducteur ?

**Réponse :**

Estimer la charge cognitive d'un conducteur est utile pour plusieurs raisons. D'abord pour la sécurité routière : si un conducteur est mentalement surchargé, ses performances se dégradent (temps de réaction plus lent, mauvaise prise de décision), ce qui augmente le risque d'accident. Ensuite, cela ouvre la voie aux véhicules intelligents adaptatifs : un système embarqué capable de détecter une surcharge cognitive pourrait automatiquement simplifier l'interface, réduire les distractions ou déclencher une alerte. Enfin, ça sera utile pour concevoir des interfaces homme-machine plus ergonomiques et adaptées à l'état réel du conducteur.

## 2. Définition : charge cognitive

#### Q2. Comment définir la charge cognitive ?

**Réponse :**

La charge cognitive désigne la quantité de ressources mentales mobilisées par la mémoire de travail pour accomplir une tâche à un instant donné. En d'autres termes, c'est l'effort mental requis pour traiter les informations disponibles et prendre des décisions. On distingue deux types : la charge intrinsèque (liée à la complexité inhérente de la tâche) et la charge extrinsèque (liée aux facteurs environnementaux non pertinents pour la tâche).

#### Q3. Dans le contexte de la conduite, de quels facteurs la charge cognitive peut-elle dépendre ?

**Réponse :**
En conduite, la charge cognitive dépend de nombreux facteurs :

La complexité de la route (virages serrés, conditions météo difficiles, trafic dense)
Les conditions environnementales (nuit, neige, visibilité réduite)
Les tâches secondaires (téléphone, GPS, conversation)
L'expérience du conducteur (un conducteur novice sera plus chargé cognitivement sur une même route)
La vitesse et la nécessité de réagir rapidement à des événements imprévus

Dans le papier, la charge cognitive est évaluée à l'aide d'une échelle subjective appelée **PAAS**, allant de 1 à 9.

#### Q5. Pourquoi parle-t-on ici d'une vérité terrain subjective ?

**Réponse :**

Parce que les scores PAAS sont auto-rapportés par les participants eux-mêmes. Il n'existe pas de mesure objective et universelle de la charge cognitive : deux individus soumis au même scénario peuvent ressentir et déclarer des niveaux différents. De plus, la perception de sa propre charge cognitive peut varier selon l'état émotionnel, la fatigue ou les habitudes de chaque personne. Cela introduit donc une part de variabilité et d'imprécision.

## 5. Modalités mesurées

Le dataset CL-Drive est **multimodal**. Il contient plusieurs types de signaux enregistrés pendant la même tâche de conduite.

#### Q7. Compléter le tableau suivant.

| Modalité | Signification | Signal mesuré | Rôle possible |
|---|---|---|---|
| EEG | Électroencéphalographie | Activité électrique du cerveau (4 canaux : AF7, AF8, TP9, TP10) à 256 Hz |Modalité principale, capte directement l'activité cérébrale liée à la charge cognitive |
| ECG | Électrocardiographie | Activité électrique du cœur à 512 Hz| Mesure la variabilité du rythme cardiaque, indicateur du stress et de l'effort mental |
| EDA | Activité électrodermale | Conductance de la peau à 128 Hz |Reflète l'activation du système nerveux sympathique, liée à l'arousal et au stress |
| Gaze | Suivi oculaire | Mouvements des yeux (saccades, fixations, diamètre pupillaire, clignements) à 50 Hz |Le diamètre pupillaire et les patterns de fixation sont corrélés à la charge cognitive |

Dans le projet, la modalité principale sera l’**EEG**. Les autres modalités serviront éventuellement d’extensions pour discuter de la fusion multimodale.

#### Q8. Pourquoi peut-il être intéressant de combiner EEG, ECG, EDA et Gaze pour estimer la charge cognitive ?

**Réponse :**

Parce que chaque modalité capture une dimension différente de l'état cognitif :

L'EEG reflète directement l'activité cérébrale mais est très bruité et variable entre individus
L'ECG capte la réponse du système cardiovasculaire à l'effort mental
L'EDA mesure la réponse du système nerveux sympathique, liée à l'éveil et au stress
Le Gaze reflète où et comment le conducteur dirige son attention

En les combinant, on obtient une image plus complète et plus robuste de l'état cognitif. C'est ce que confirment les résultats du papier : les modèles multimodaux sont systématiquement meilleurs que les modèles unimodaux.

## 7. Protocole expérimental

#### Q11. Quelle est la différence entre un **protocole expérimental** et un **dataset exploitable en machine learning** ?

**Réponse :**

Un protocole expérimental c'est l'ensemble des conditions et étapes mises en place pour collecter les données : qui sont les participants, quelles tâches ils réalisent, quels capteurs on utilise, dans quel ordre se déroule la session, etc. C'est la méthodologie de collecte.
Un dataset exploitable en ML c'est le résultat après traitement de ces données brutes : les signaux sont nettoyés, segmentés en fenêtres, les features sont extraites, et chaque fenêtre est associée à un label. C'est ce qu'on va réellement donner à manger à un modèle.

#### Q13. Pourquoi les auteurs collectent-ils une baseline avant les scénarios ?

**Réponse :**

La baseline sert de référence individuelle pour chaque participant. Les signaux physiologiques comme l'EEG ou l'EDA varient naturellement d'une personne à l'autre, même au repos. En enregistrant l'état de base de chaque participant, on peut ensuite normaliser ses signaux par rapport à lui-même plutôt que par rapport à une moyenne générale. Cela réduit la variabilité inter-sujets et améliore la qualité des données pour le ML.

#### Q14. Pourquoi prévoir un scénario d’orientation avant les scénarios principaux ?

**Réponse :**

Pour deux raisons principales. D'abord pour que le participant s'habitue au simulateur et réduire le risque de SAS dès le début. Ensuite pour qu'il calibre son échelle PAAS : en voyant un scénario simple (autoroute), il comprend ce que signifie une charge faible, et peut mieux évaluer les niveaux suivants par comparaison.

#### Q15. Pourquoi prévoir des périodes de repos entre les scénarios ?

### Réponse

Les périodes de repos de 2 minutes servent à trois choses :

Laisser le participant récupérer mentalement pour que chaque scénario parte d'un état cognitif similaire
Réduire le risque de SAS qui peut s'accumuler au fil des scénarios
Éviter l'effet de carry-over : sans repos, la charge cognitive d'un scénario difficile pourrait contaminer les mesures du scénario suivant

#### Q16. Pourquoi demander un score de charge cognitive toutes les 10 secondes plutôt qu’une seule fois à la fin du scénario ?

### Réponse

Parce qu'un score unique en fin de scénario poserait deux problèmes majeurs. D'abord le participant oublie ou moyenne ce qu'il a ressenti au début du scénario, le score final est donc peu fiable. Ensuite, on n'aurait qu'un seul label pour 3 minutes de signal, ce qui est beaucoup trop peu pour entraîner un modèle ML correctement. Avec 18 labels par scénario, on obtient des paires signal/label beaucoup plus précises et exploitables.

## 8. Scénarios de conduite

Le dataset contient 9 scénarios principaux, conçus pour induire différents niveaux de charge cognitive.

#### Q17. Compléter le tableau suivant.

| Scénario | Description synthétique |
|---:|---|
| 1 |Conduite sur autoroute de jour, maintenir 80 km/h en ligne droite — charge très faible |
| 2 | Conduite de nuit à 80 km/h — même route mais visibilité réduite, légère augmentation de la charge |
| 3 | Conduite de nuit sur autoroute avec neige à 80 km/h — conditions dégradées, charge modérée |
| 4 | Tennis ball challenge — toucher une balle avec le pneu en maintenant précision et vitesse — début des défis de précision |
| 5 | Slalom — naviguer entre des portes en maintenant précision et vitesse — coordination plus exigeante |
| 6 | Passage étroit — traverser des couloirs étroits avec précision tout en accélérant — charge croissante |
| 7 | Virages à 90° — enchaîner des virages gauche/droite à 90° rapidement avec précision — charge élevée |
| 8 | Demi-tour en 3 points — manœuvre complexe sur route étroite en suivant des instructions vocales — très charge élevée |
| 9 | Ruelle étroite — naviguer dans une ruelle très étroite avec précision et vitesse — charge maximale |

L'hypothèse générale est que les scénarios les plus complexes induisent une charge cognitive plus élevée.

#### Q18. Pourquoi utiliser plusieurs scénarios de conduite au lieu d'un seul scénario ?

**Réponse :**

Pour deux raisons principales :
1. Éviter l'effet d'apprentissage inversé
Si on commençait par les scénarios très difficiles, le participant serait désorienté et ses scores PAAS seraient faussés par la surprise et le stress de la nouveauté. En commençant simple, il monte progressivement en difficulté de façon naturelle.
2. Cohérence des labels PAAS
Si le participant a déjà vécu un scénario difficile, il peut mieux calibrer son ressenti sur les scénarios suivants. L'ordre croissant lui donne une référence progressive pour évaluer sa propre charge.

#### Q19. Pourquoi peut-on s’attendre à une charge cognitive plus élevée dans les scénarios 8 et 9 ?

### Réponse

Plusieurs raisons: fin des tests donc possiblement fatigue accumulé à la suite de toutes les expériences. 
Ces scénarios sont assez complexe et représente donc une potentiel augmentation de charge cognitive en comparaison des scénarios précédants.

#### Q20. La complexité du scénario suffit-elle à définir le label de charge cognitive ?

### Réponse

Non, et c'est précisément l'intérêt du dataset.

La complexité du scénario est objective : le scénario 8 est intrinsèquement plus difficile que le scénario 1, c'est conçu ainsi par les auteurs.
Mais la charge cognitive est subjective : elle dépend de la personne qui conduit. Un conducteur expérimenté pourrait trouver le scénario 5 peu chargé cognitivement, là où un conducteur moins habitué le vivrait comme très stressant.

## 10. Formulation du problème d'apprentissage automatique

À partir du dataset CL-Drive, on peut formuler plusieurs problèmes de classification.

### Classification binaire

| Scores PAAS | Classe |
|---|---|
| 1 à 4 | charge faible |
| 5 à 9 | charge élevée |

### Classification ternaire

| Scores PAAS | Classe |
|---|---|
| 1 à 3 | charge faible |
| 4 à 6 | charge moyenne |
| 7 à 9 | charge élevée |

Dans le projet, on commencera par la classification binaire.

#### Q27. Pourquoi commencer par une classification binaire plutôt qu’une classification ternaire ?

### Réponse

La classification binaire est plus simple car on ne distingue que deux classes (charge faible vs élevée), ce qui réduit la difficulté du problème et permet d'obtenir de meilleures performances initiales. C'est une bonne façon de valider que les signaux contiennent bien de l'information utile avant de passer à un problème plus difficile à 3 classes.

#### Q28. Dans une classification binaire, pourquoi le score PAAS 5 est-il généralement placé dans la classe « charge élevée » dans ce papier ?

### Réponse

Le score 5 signifie "ni faible, ni élevée" sur l'échelle PAAS, donc il est ambigu. Les auteurs ont choisi de le placer dans la classe élevée car dans un contexte de sécurité routière, il vaut mieux surestimer la charge cognitive que la sous-estimer. Classer un conducteur en surcharge alors qu'il ne l'est pas vraiment est moins dangereux que l'inverse.

#### Q29. Quel est le lien entre ce papier et un système d'interface adaptative ?

**Réponse :**

Ce papier est directement lié aux interfaces adaptatives : si un véhicule intelligent peut détecter en temps réel que le conducteur est en surcharge cognitive, il peut adapter automatiquement son comportement, par exemple en simplifiant l'affichage du tableau de bord, en suspendant les notifications, en activant une alerte, ou en proposant une assistance à la conduite. Le modèle ML entraîné sur CL-Drive est le cœur de ce système de détection.

## 9. Visualisation des signaux EEG

Avant tout prétraitement ou apprentissage automatique, il est indispensable de visualiser les signaux EEG.

La visualisation permet de :

- vérifier que les données sont correctement chargées ;
- observer les amplitudes ;
- repérer les valeurs aberrantes ;
- détecter les ruptures ou données manquantes ;
- comparer les canaux ;
- repérer des artefacts évidents ;
- comparer qualitativement des segments de faible et forte charge cognitive.

### Algorithme conceptuel pour visualiser un segment EEG

Pour visualiser un segment EEG de 10 secondes :

1. choisir un sujet ;
2. choisir un scénario ;
3. choisir un segment ;
4. récupérer les 4 canaux EEG ;
5. calculer le nombre d’échantillons : $N = f_s \times T$ ;
6. construire l’axe temporel : $t[n] = n/f_s$ ;
7. tracer les quatre canaux en fonction du temps ;
8. ajouter le score PAAS associé au titre du graphique ;
9. commenter la qualité du signal.

#### Q30. Que faut-il regarder en premier lorsqu’on affiche un signal EEG brut ?

### Réponse

En premier, il faut regarder l'amplitude générale du signal : est-elle dans une plage normale (typiquement entre -100 et +100 µV pour l'EEG) ? Ensuite on vérifie l'absence de valeurs aberrantes (pics extrêmes), de segments plats (signal mort ou déconnexion), et de ruptures brutales qui signaleraient un artefact ou une perte de signal.

#### Q31. Pourquoi afficher les quatre canaux EEG plutôt qu’un seul ?

### Réponse

Parce que chaque canal capte une région différente du cerveau. Afficher les 4 permet de détecter si un problème est global (tous les canaux affectés → problème matériel ou artefact de mouvement) ou local (un seul canal affecté → mauvais contact d'une électrode). De plus, la charge cognitive peut se manifester différemment selon les régions frontales et temporales.

#### Q32. Pourquoi comparer visuellement un segment de faible charge et un segment de forte charge cognitive ?

### Réponse

Pour valider visuellement que les signaux sont exploitables. Si on observe des différences visuelles entre les deux (amplitude, fréquence, irrégularité), cela confirme que l'EEG capture bien quelque chose de différent entre les deux états cognitifs. Si les signaux se ressemblent parfaitement, cela peut indiquer un problème de qualité des données.

#### Q33. Pourquoi l'utilisation d'un simulateur est-elle à la fois un avantage et une limite ?

**Réponse :**

Avantages :

Environnement totalement contrôlé : on peut reproduire exactement les mêmes conditions pour tous les participants (météo, trafic, obstacles)
Sécurité totale : pas de risque d'accident réel pendant l'expérience

Limites :

Risque de SAS (Simulation Adaptation Syndrome) : certains participants développent des malaises, ce qui a forcé l'arrêt de 2 sessions
L'expérience ne reproduit pas parfaitement la conduite réelle : absence de vrais risques, graphismes artificiels, sensations limitées → la charge cognitive ressentie peut être différente de celle en situation réelle

Le papier CL-Drive s'intéresse à l'estimation automatique de la charge cognitive d'un conducteur à partir de signaux physiologiques mesurés pendant la conduite. L'expérience a été menée avec 21 participants conduisant dans un simulateur de véhicule immersif, dans 9 scénarios de complexité croissante durant chacun 3 minutes. Quatre types de signaux ont été enregistrés en simultané : l'EEG (activité cérébrale), l'ECG (activité cardiaque), l'EDA (conductance cutanée) et le Gaze (mouvements oculaires). Toutes les 10 secondes, les participants déclaraient verbalement leur niveau de charge cognitive sur l'échelle PAAS allant de 1 à 9, constituant ainsi la vérité terrain. Ces scores subjectifs sont qualifiés de vérité terrain subjective car ils reposent sur le ressenti propre de chaque individu, qui peut varier d'une personne à l'autre pour un même scénario. Le problème est formulé comme une tâche de classification : binaire (charge faible vs élevée) ou ternaire (faible / moyenne / élevée), à partir de fenêtres de signal de 10 secondes. Les meilleurs résultats atteignent 83% de précision en classification binaire avec le modèle XGBoost utilisant les 4 modalités. L'intérêt applicatif est direct : un tel système pourrait être intégré dans un véhicule intelligent pour détecter en temps réel la surcharge cognitive du conducteur et adapter automatiquement l'interface, sans que celui-ci ait besoin de déclarer son état.

## Activité 4 - Synthèse personnelle

1. **Ce que j'ai compris du papier :**

Le papier CL-Drive propose un dataset multimodal pour estimer automatiquement la charge cognitive d'un conducteur à partir de signaux physiologiques (EEG, ECG, EDA, Gaze) enregistrés pendant une conduite simulée. Le protocole repose sur 9 scénarios de complexité croissante, avec des labels subjectifs PAAS collectés toutes les 10 secondes. La tâche est formulée comme une classification binaire (charge faible vs élevée) ou ternaire. Le modèle XGBoost appliqué aux features manuelles obtient les meilleurs résultats en 10-fold (83,67 % en binaire avec les 4 modalités), tandis que les réseaux VGG/ResNet dominent en LOSO. L'ensemble confirme que l'EEG seul est moins performant que la combinaison des 4 modalités, et que la généralisation inter-sujets (LOSO) reste nettement plus difficile que la validation intra-sujets (10-fold).

2. **Ce qui me semble le plus difficile :**

La variabilité inter-sujets est clairement le point le plus délicat : les signaux EEG varient énormément d'une personne à l'autre, ce qui explique l'écart important entre les scores 10-fold (~83 %) et LOSO (~76 %). Par ailleurs, la vérité terrain repose sur des scores auto-rapportés (PAAS), ce qui introduit du bruit dans les labels. Enfin, la segmentation en fenêtres de 10 secondes et l'association correcte de chaque fenêtre à son label constituent une étape technique délicate à implémenter sans erreur de décalage temporel.
